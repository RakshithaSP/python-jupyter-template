### Exercise 1: Line Plot Customization

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.linspace(0,10,100)
y=x**2
plt.plot(x,y)
plt.title('Line plot of y= x^2')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```
