### Exercise 2: Bar Plot with Colors

```python
import matplotlib.pyplot as plt
import numpy as np
categories=['A','B','C','D']
values=[5,7,3,9]
plt.bar(categories,values,color=['red','green','blue','yellow'])
plt.title('Bar plot Example')
plt.xlabel('category')
plt.ylabel('value')
plt.show()
```
